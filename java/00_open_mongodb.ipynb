# Open MongoDB Using the MongoDB Extension

1. click on the MongoDB extension on the side panel
2. To connect to the local database, click on ‘Library DB’ in the list of connections on the left side

You should see four databases in the MongoDB database server:
- admin
- config
- local
- library -> __we're going to work with this one__

## Open the library database

1. Open the library database
1. Open the books collection
1. Click on Documents
1. Open a document

## Connecting to Atlas (or your own deployment)

- add a new connection clicking on `+` (`Add MongoDB Connection`).
- enter the [MongoDB Connection String](https://www.mongodb.com/docs/manual/reference/connection-string/) of your database.
